```python
#Importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
# Select independent and dependent variable

X = dataset.iloc[:,2:32]
y = dataset.iloc[:,1]
```

```python
X
```

```python
# Feature Scaling (standardization)

from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
X
```

```python
# Data preprocessing (Encoding)
from sklearn.preprocessing import LabelEncoder
y=LabelEncoder().fit_transform(y)
y
```

```python
#Splitting the dataset into the training 70% and test 30 dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
print("X_train shape : ",X_train.shape)
print("X_test shape : ",X_test.shape)
print("y_train shape : ",y_train.shape)
print("y_test shape : ",y_test.shape)
```

```python
# SVM Model Building and training

from sklearn.svm import SVC
classifier = SVC()

start = time.time()
classifier.fit(X_train,y_train)
end = time.time()
Final  = end - start
print(str(Final) + 'Second')
```

```python
# Model Prediction

start = time.time()
y_pred=classifier.predict(X_test)
end = time.time()
Final = end - start
print(str(Final) + 'Second')
```

```python
from sklearn.metrics import hinge_loss
hinge_loss(y_test, y_pred)
```

```python
# Display Performance analyses 

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_auc_score
from imblearn.metrics import specificity_score

print("Accuracy score : ",accuracy_score(y_test,y_pred))
print("")
print("Specificity score : ",specificity_score(y_test,y_pred))
print("")
print("Roc Auc score : ",roc_auc_score(y_test,y_pred))
print("")
print(classification_report(y_test,y_pred))
print("")
print(confusion_matrix(y_test,y_pred))
print("")
print("Heatmap report : ",sns.heatmap(confusion_matrix(y_test,y_pred),annot=True))
```

```python
# Hyper parameter tunning (grid searchcv)

param_grid={
    'C':[.1,1,10,100,1000],
    'kernel':['rbf','poly'],
    'gamma':[1,.1,.01,.001,.0001]
}
```

```python
# Grid searchcv building and training

from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(SVC(),param_grid,cv=5, verbose=3).fit(X_train,y_train)
```

```python
# Collect Best parameter and best score

print('Best parameter : ', grid_search.best_params_)
print('Best score : ', grid_search.best_score_)
```

```python
# SVM model building and training

from sklearn.svm import SVC
tunning_cfr = SVC(kernel='rbf',C=10,gamma=0.01).fit(X_train,y_train)
```

```python
# Model Prediction
tunning_pred= tunning_cfr.predict(X_test)
```

```python
from sklearn.metrics import hinge_loss
hinge_loss(y_test, tunning_pred)
```

```python
print("Accuracy score : ",accuracy_score(y_test,tunning_pred))
print("")
print("Specificity score : ",specificity_score(y_test,tunning_pred))
print("")
print("Roc auc score : ",roc_auc_score(y_test,tunning_pred))
print(classification_report(y_test,tunning_pred))
print("")
print(confusion_matrix(y_test,tunning_pred))
print("")
print(sns.heatmap(confusion_matrix(y_test,tunning_pred),annot=True))
```
